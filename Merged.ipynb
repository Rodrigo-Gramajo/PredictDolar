```python
def remove_large_values(df, columns, max_value=np.finfo(np.float64).max):
    for col in columns:
        df[col] = df[col].apply(lambda x: max_value if x > max_value else x)
    return df

df = remove_large_values(df, df.columns)
```

```python
from sklearn.preprocessing import StandardScaler

columns = ['Año', 'Consumo Cerveza Per Capita', 'Consumo Fertilizante',
       'Consumo Pescados Per Capita', 'Consumo Recursos Fosiles Per Capita',
       'Consumo Fruta Per Capita', 'Consumo Gas', 'Consumo Aceite Per Capita',
       'Consumo Huevos Per Capita', 'Consumo Leche Per Capita',
       'Consumo Vegetales Per Capita', 'Consumo Vino Per Capita',
       'Exportacion Bienes & Servicios', 'Exportacion de Comida',
       'Exportacion de Manufacturas', 'Exportacion de Servicios',
       'Gasto en Salud', 'Gasto en Educacion', 'Tasa de Natalidad',
       'Tasa de Mortalidad', 'Muerte Infantil', 'Expectativa de Vida',
       'Suscripciones a Celulares', 'Utilizacion de Internet',
       'Densidad Poblacional', 'Produccion Agricola $', 'Produccion de Carne',
       'Produccion de Leche', 'PBI per Capita']

scaler = StandardScaler()
df[columns] = scaler.fit_transform(df[columns])
```

```python
# Training and Testing the data
train, test = train_test_split(df, test_size=0.2, shuffle=True)
training_features = ['Año', 'Consumo Cerveza Per Capita', 'Consumo Fertilizante',
       'Consumo Pescados Per Capita', 'Consumo Recursos Fosiles Per Capita',
       'Consumo Fruta Per Capita', 'Consumo Gas', 'Consumo Aceite Per Capita',
       'Consumo Huevos Per Capita', 'Consumo Leche Per Capita',
       'Consumo Vegetales Per Capita', 'Consumo Vino Per Capita',
       'Exportacion Bienes & Servicios', 'Exportacion de Comida',
       'Exportacion de Manufacturas', 'Exportacion de Servicios',
       'Gasto en Salud', 'Gasto en Educacion', 'Tasa de Natalidad',
       'Tasa de Mortalidad', 'Muerte Infantil', 'Expectativa de Vida',
       'Suscripciones a Celulares', 'Utilizacion de Internet',
       'Densidad Poblacional', 'Produccion Agricola $', 'Produccion de Carne',
       'Produccion de Leche']
target = 'PBI per Capita'
train_X = train[training_features]
train_Y = train[target]
test_X = test[training_features]
test_Y = test[target]
```

```python
# linear regression model
model = LinearRegression()
model.fit(train_X, train_Y)
train_pred_Y = model.predict(train_X)
test_pred_Y = model.predict(test_X)
train_pred_Y = pd.Series(train_pred_Y.clip(0, train_pred_Y.max()), index=train_Y.index)
test_pred_Y = pd.Series(test_pred_Y.clip(0, test_pred_Y.max()), index=test_Y.index)

from sklearn.metrics import mean_absolute_error

# calculate mean absolute error
mae_train = mean_absolute_error(train_Y, train_pred_Y)
mae_test = mean_absolute_error(test_Y, test_pred_Y)

# print results
print('Mean Absolute Error (train):', mae_train)
print('Mean Absolute Error (test):', mae_test)
```
